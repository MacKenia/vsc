# 随机出题

```python
import sys
import random
from PySide6.QtWidgets import (QApplication, QPushButton, QHBoxLayout, QVBoxLayout,
        QWidget, QLineEdit, QLabel, QListWidget)
from PySide6.QtCore import Qt

class randYourSelf(QWidget):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("做尼玛")

        self.sum = [0,0]

        self.quiz = []
        self.quiz.append(QLabel())
        self.quiz.append(QLabel())
        self.quiz.append(QLabel())
        self.quiz.append(QLabel("="))
        self.quiz.append(QLineEdit())

        self.quiz[0].setAlignment(Qt.AlignCenter)
        self.quiz[2].setAlignment(Qt.AlignCenter)
        self.quiz[4].setAlignment(Qt.AlignCenter)

        self.pl1 = QHBoxLayout()
        for i in self.quiz:
            self.pl1.addWidget(i)

        self.quiz[4].returnPressed.connect(self.confirm)

        self.button = QPushButton("确认")
        self.button.clicked.connect(self.confirm)


        self.history = QListWidget()
        self.pl2 = QVBoxLayout()

        self.s = QLabel(f"总题数: {self.sum[0] + self.sum[1]}  🟢: {self.sum[0]}\t🔴: {self.sum[1]}\t正确率: 0%")

        self.pl2.addLayout(self.pl1)
        self.pl2.addWidget(self.button)
        self.pl2.addWidget(self.history)
        self.pl2.addWidget(self.s)

        self.setLayout(self.pl2)
        self.update()
        self.show()

    def confirm(self):
        if self.t[3] == int(self.quiz[4].text()):
            self.sum[0] += 1
            self.history.addItem("{} {} {} = {}\t✔".format(self.t[0], self.t[1], self.t[2], self.t[3]))
        else:
            self.sum[1] += 1
            self.history.addItem("{} {} {} = {}\t✘".format(self.t[0], self.t[1], self.t[2], self.quiz[4].text()))
        self.s.setText(f"总题数: {self.sum[0] + self.sum[1]}  🟢: {self.sum[0]}\t🔴: {self.sum[1]}\t正确率: {self.sum[0] / (self.sum[0] + self.sum[1]) * 100:.0f}%")
        self.update()
        self.quiz[4].setText("")
        return
    
    def update(self):
        self.t = self.gen()
        for i, x in enumerate(self.t):
            if i == 3:
                break
            self.quiz[i].setText(str(x))


    def gen(self):
        # 随机出题系统 10以内四则运算 并统计正确率
        a = b = n = 0
        op = ["+", "-", "*", "/"]
        sum = [0, 0]

        o = random.randint(0, 3)
        a = random.randint(1, 10)
        b = random.randint(1, 10)
         
        # 排除特殊情况
        if o == 1:
            if a < b:
                a, b = b, a
        if o == 3:
            while a % b != 0 or a < b:
                a = random.randint(1, 10)
                b = random.randint(1, 10)

        ans = 0

        if o == 0:
            ans = a + b
        elif o == 1:
            ans = a - b
        elif o == 2:
            ans = a * b
        elif o ==3:
            ans = a / b

        return (a, op[o], b, ans)

if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    
    ran = randYourSelf()
    
    app.exec()
```

# 登录升级

```python
import sys
from PySide6.QtWidgets import (QApplication, QPushButton, QHBoxLayout, QVBoxLayout, QLineEdit, QLabel, QDialog)
from PySide6.QtCore import Qt


from PySide6.QtCore import Qt

class Form(QDialog):

    def __init__(self, parent=None):
        super(Form, self).__init__(parent)
        # Create widgets
        self.user = QLabel("用户名: ")
        self.passwd = QLabel("密   码: ")
        self.user_input = QLineEdit()
        self.passwd_input = QLineEdit()
        self.button = QPushButton("登录")
        self.errs = QLabel()
        # Create layout and add widgets
        self.main_layout = QVBoxLayout()
        self.user_layout = QHBoxLayout()
        self.passwd_layout = QHBoxLayout()

        self.user_layout.addWidget(self.user)
        self.user_layout.addWidget(self.user_input)

        self.passwd_layout.addWidget(self.passwd)
        self.passwd_layout.addWidget(self.passwd_input)

        self.main_layout.addLayout(self.user_layout)
        self.main_layout.addLayout(self.passwd_layout)
        self.main_layout.addWidget(self.button)
        self.main_layout.addWidget(self.errs)
        # Set dialog layout
        self.setLayout(self.main_layout)
        # Add button signal to greetings slot
        self.button.clicked.connect(self.greetings)
        self.err = 0

    # Greets the user
    def greetings(self):
        if self.user_input.text() == "user1" and self.passwd_input.text() == "123":
            dialog = QDialog()
            label = QLabel("欢迎你, user1")
            label.setAlignment(Qt.AlignCenter)
            layout = QVBoxLayout()
            layout.addWidget(label)
            dialog.setLayout(layout)
            dialog.exec()
        else:
            self.err += 1
            if self.err == 3:
                self.button.setDisabled(True)
                self.errs.setText("错误次数过多，请稍后再试")
            else:
                self.errs.setText(f"错误次数: {self.err}")
            self.errs.setAlignment(Qt.AlignCenter)
            self.errs.setStyleSheet("color: red")
            self.errs.show()

if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    form = Form()
    form.show()
    app.exec()
```

# 单选框

```python
import sys
from PySide6.QtWidgets import (QApplication, QDialog, QVBoxLayout, QHBoxLayout, QRadioButton, QLabel, QPushButton, QGroupBox)
from PySide6.QtCore import Qt

class wan(QDialog):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("来！猜！！！")
        self.question = QLabel("老师让小猫和小狗去背书，请问谁先背书呢？")
        self.c1 = QRadioButton("小猫")
        self.c2 = QRadioButton("小狗")
        self.button = QPushButton("提交")
        self.ans = QLabel()

        self.radioG = QGroupBox()
        self.radioG.setLayout(QVBoxLayout())
        self.radioG.layout().addWidget(self.c1)
        self.radioG.layout().addWidget(self.c2)
        self.radioG.setFlat(True)

        self.main_layout = QVBoxLayout()
        self.main_layout.addWidget(self.question)
        self.main_layout.addWidget(self.radioG)
        self.main_layout.addWidget(self.button)
        self.main_layout.addWidget(self.ans)
        self.setLayout(self.main_layout)
        self.button.clicked.connect(self.check)

    def check(self):
        if self.c1.isChecked():
            self.ans.setText("错误，答案是小狗，因为旺旺仙贝(汪汪先背)")
        else:
            self.ans.setText("回答正确,旺旺仙贝(汪汪先背)")

if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    w = wan()
    w.show()
    app.exec()
```

# 用户注册界面

```python
import sys
from PySide6.QtWidgets import (QApplication, QDialog, QVBoxLayout, QHBoxLayout, QGridLayout, QRadioButton,
        QLabel, QPushButton, QGroupBox, QCheckBox)
from PySide6.QtCore import Qt

class sign_in(QDialog):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("注册")
        self.id = QLabel("学号: ")
        self.user = QLabel("姓名: ")
        self.gander = QGroupBox("性别: ")
        self.hobby = QGroupBox("爱好: ")
        
        self.id_input = QLineEdit()
        self.user_input = QLineEdit()
        
        self.male = QRadioButton("男")
        self.female = QRadioButton("女")

        self.hobbes = ["唱歌", "跳舞", "篮球", "坐敞篷车", "钓鱼"]
        self.hobbes_ = []

        self.id_box = QHBoxLayout()
        self.id_box.addWidget(self.id)
        self.id_box.addWidget(self.id_input)

        self.user_box = QHBoxLayout()
        self.user_box.addWidget(self.user)
        self.user_box.addWidget(self.user_input)

        self.gander_box = QHBoxLayout()
        self.gander_box.addWidget(self.male)
        self.gander_box.addWidget(self.female)
        self.gander.setLayout(self.gander_box)

        self.hobby_box = QGridLayout()
        self.hobby_box.setColumnMinimumWidth(3, 0)
        for i in self.hobbes:
            self.hobbes_.append(QCheckBox(i))
            self.hobby_box.addWidget(self.hobbes_[len(self.hobbes_)-1])
        self.hobby.setLayout(self.hobby_box)

        self.button = QPushButton("提交")
        self.button.clicked.connect(self.shows)

        self.main_layout = QVBoxLayout()
        self.main_layout.addLayout(self.id_box)
        self.main_layout.addLayout(self.user_box)
        self.main_layout.addWidget(self.gander)
        self.main_layout.addWidget(self.hobby)
        self.main_layout.addWidget(self.button)
        self.setLayout(self.main_layout)
        
    def shows(self):
        log = QDialog()
        log.setWindowTitle("注册成功")
        info = []
        info.append("学号: "+self.id_input.text())
        info.append("姓名: "+self.user_input.text())
        if self.male.isChecked():
            info.append("性别: 男")
        elif self.female.isChecked():
            info.append("性别: 女")
        info.append("爱好:")
        for i in self.hobbes_:
            if i.isChecked():
                info.append("\t"+i.text())

        layout = QVBoxLayout()
        for i in info:
            layout.addWidget(QLabel(i))
        log.setLayout(layout)
        log.exec()


if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    w = sign_in()
    w.show()
    app.exec()
```

# 选课器

```python
import sys
from PySide6.QtWidgets import (QApplication, QDialog, QVBoxLayout, QHBoxLayout, QRadioButton, QListWidget, QPushButton)
from PySide6.QtCore import Qt

class course(QDialog):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("课程")

        self.courses = ["Python 程序设计", "数据结构", "计算机网络", "Java 程序设计"]

        self.course_list_remain = QListWidget()
        self.course_list_choose = QListWidget()
        self.confirm = QPushButton(">>")
        self.cancel = QPushButton("<<")

        for i in self.courses:
            self.course_list_remain.addItem(i)

        self.buttons = QVBoxLayout()
        self.buttons.addWidget(self.confirm)
        self.buttons.addWidget(self.cancel)

        self.confirm.setFixedWidth(50)
        self.cancel.setFixedWidth(50)

        self.main = QHBoxLayout()
        self.main.addWidget(self.course_list_remain)
        self.main.addLayout(self.buttons)
        self.main.addWidget(self.course_list_choose)

        self.setLayout(self.main)

        self.resize(300, 300)

        self.confirm.clicked.connect(self.f_confirm)
        self.cancel.clicked.connect(self.f_cancel)

    def f_confirm(self):
        for i in range(self.course_list_remain.count()):
            if self.course_list_remain.item(i).isSelected():
                self.course_list_choose.addItem(self.course_list_remain.item(i).text())
                self.course_list_remain.takeItem(i)
        pass

    def f_cancel(self):
        for i in range(self.course_list_choose.count()):
            if self.course_list_choose.item(i).isSelected():
                self.course_list_remain.addItem(self.course_list_choose.item(i).text())
                self.course_list_choose.takeItem(i)
        pass

if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    w = course()
    w.show()
    app.exec()
```

```python
import sys
from PySide6.QtWidgets import (QApplication, QDialog, QHBoxLayout, QLabel, QPushButton, QComboBox)
from PySide6.QtCore import Qt

class vip(QDialog):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("VIP")

        self.vip_ = QLabel("选择会员:")
        self.menu = QComboBox()
        self.vip = QPushButton("一键选择钻石会员")
        self.submit = QPushButton("提交")
        self.tip = QLabel()

        self.menu.addItem("38垃圾会员")
        self.menu.addItem("68普通会员")
        self.menu.addItem("128钻石会员")

        self.up = QHBoxLayout()
        self.up.addWidget(self.vip_)
        self.up.addWidget(self.menu)
        self.up.addWidget(self.vip)

        self.main = QVBoxLayout()
        self.main.addLayout(self.up)
        self.main.addWidget(self.submit, 0, Qt.AlignCenter)
        self.submit.setFixedWidth(100)
        self.main.addWidget(self.tip, 0, Qt.AlignCenter)

        self.setLayout(self.main)

        self.resize(300, 100)

        self.submit.clicked.connect(self.f_submit)
        self.vip.clicked.connect(self.f_vip)

    def f_submit(self):
        if self.menu.currentText() == "128钻石会员":
            self.tip.setText("恭喜！128元钻石会员办理成功")
            self.tip.setStyleSheet("color:#ff0000")

    def f_vip(self):
        self.menu.setCurrentText("128钻石会员")

if __name__ == "__main__":
    app = QApplication.instance()
    if not app:
        app = QApplication(sys.argv)
    w = vip()
    w.show()
    app.exec()
        
```
